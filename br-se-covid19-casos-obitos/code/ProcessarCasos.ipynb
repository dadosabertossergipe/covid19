# Dados de Casos e Óbitos

Os dados de casos e óbitos são filtrados a partir dos dados disponibilizados pelo site [brasil.io](https://brasil.io). Os dados mais atualizados são apenas do estado como um todos. Os dados dos municípios não estão atualizados. 

O arquivo de `input` não está incluso nesse repositório por conta do tamanho. 

```python
import pandas as pd
```

```python
dados_vacinacao = pd.read_csv("../input/brasilio/caso_full.csv")
```

```python
dados_sergipe = dados_vacinacao.query("state == 'SE' and place_type == 'state'").sort_values(by='date')
dados_sergipe = dados_sergipe[['date','epidemiological_week','last_available_confirmed','last_available_deaths','new_confirmed','new_deaths']]
dados_sergipe = dados_sergipe.reset_index(drop=True)
dados_sergipe['mean7_confirmed'] = dados_sergipe['new_confirmed'].rolling(window=7).mean()
dados_sergipe['mean7_deaths'] = dados_sergipe['new_deaths'].rolling(window=7).mean()
dados_sergipe
```

```python
dados_sergipe.columns = ['data','semana_epidemiologica','total_casos_confirmados','total_obitos_confirmados','diario_casos_confirmados','diario_obitos_confirmados','media_movel7_casos','media_movel7_obitos']
dados_sergipe
```

```python
dados_sergipe.to_csv("../output/se_covid19_casos_obitos.csv", index=False)
```
