# Dados de Casos e Óbitos

Os dados de casos e óbitos são filtrados a partir dos dados disponibilizados pelo site [brasil.io](https://brasil.io). Os dados mais atualizados são apenas do estado como um todos. Os dados dos municípios não estão atualizados. 

O arquivo de `input` não está incluso nesse repositório por conta do tamanho. 

```python
import pandas as pd
```

```python
dados_vacinacao = pd.read_csv("../input/brasilio/caso_full.csv")
```

```python
dados_sergipe = dados_vacinacao.query("state == 'SE' and place_type == 'state'").sort_values(by='date')
dados_sergipe = dados_sergipe[['date','epidemiological_week','last_available_confirmed','last_available_deaths','new_confirmed','new_deaths']]
dados_sergipe = dados_sergipe.reset_index(drop=True)
dados_sergipe
```

A partir do dia 27/03, os dados do brasil.io não foram mais atualizados como pode ser lido neste comunicado: https://blog.brasil.io/2022/03/28/2022-03-28-hoje-encerramos-nosso-boletim-covid19/. A partir desta data, os dados estão sendo atualizados neste repositório no arquivo: `input/casos_obitos_sergipe.xlsx`.

```python
dados_xlsx = pd.read_excel("../input/casos_obitos_sergipe.xlsx")
dados_xlsx['date'] = dados_xlsx['date'].astype(str)
dados_xlsx
```

```python
dados_sergipe_final = pd.concat([dados_sergipe, dados_xlsx], ignore_index=True)
dados_sergipe_final
```

```python
dados_sergipe_final['mean7_confirmed'] = dados_sergipe_final['new_confirmed'].rolling(window=7).mean()
dados_sergipe_final['mean7_deaths'] = dados_sergipe_final['new_deaths'].rolling(window=7).mean()
dados_sergipe_final
```

```python
dados_sergipe_final.columns = ['data','semana_epidemiologica','total_casos_confirmados','total_obitos_confirmados','diario_casos_confirmados','diario_obitos_confirmados','media_movel7_casos','media_movel7_obitos']
dados_sergipe_final
```

```python
dados_sergipe_final.to_csv("../output/se_covid19_casos_obitos.csv", index=False)
```
